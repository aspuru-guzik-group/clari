```python
from clari.inference import ClariSampler
from clari.inference.sample import sample_trajectory
from clari.chem.draw import draw_crystal_trajectory_from_batch

sampler = ClariSampler("clari-h")

out = sample_trajectory(
    sampler,
    ["CC(=O)Oc1ccccc1C(=O)O", "O"],
    copies=[1, 3],
    id="aspirin_trihydrate",
    samples=1,
)
```

```python
view = draw_crystal_trajectory_from_batch(out, batch_idx=0)
view.show()
```
